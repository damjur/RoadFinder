```python
from RoadFinder import loadImage, preprocessorX,SIZE,IMPOSITION,SIDE,load_model
import matplotlib.pyplot as plt
import numpy as np
import cv2
import keras
import os
```

```python
img = loadImage("http://www.cs.toronto.edu/~vmnih/data/mass_roads/train/sat/10078660_15.tiff")
img = cv2.resize(img,SIZE)
plt.imshow(img)
plt.show()
label = loadImage("http://www.cs.toronto.edu/~vmnih/data/mass_roads/train/map/10078660_15.tif")
label = cv2.resize(label,SIZE)
plt.imshow(label)
plt.show()
```

```python
def collectPatches(patches,shape):
    acc = np.zeros(shape)
    rowlength = shape[0]//SIDE
    for i,patch in enumerate(patches):
        y0,x0 = SIDE*(i%rowlength),SIDE * (i//rowlength)
        acc[x0:x0+SIDE,y0:y0+SIDE] = patch
        
    return acc
        

def processImage(image,model):
    patches = np.array(preprocessorX(image))
    patches = patches.reshape(len(patches),SIDE+2*IMPOSITION,SIDE+2*IMPOSITION,1)
    
    mask1 = [np.all(patch==patch.mean()) for patch in patches]
    predictions = model.predict(patches)
    for i in range(len(predictions)):
        min1 = predictions[i].min()
        d = predictions[i].max() - min1
        predictions[i] -= min1
        if d!=0:
            predictions[i] /= d
    predictions[mask1] = np.zeros_like(predictions[0])
    predictions = predictions[:,IMPOSITION:-IMPOSITION,IMPOSITION:-IMPOSITION].reshape(len(predictions),SIDE,SIDE)
    result = collectPatches(predictions,image.shape[:2])
    
    return result

patches = np.array(preprocessorX(img))[:,IMPOSITION:-IMPOSITION,IMPOSITION:-IMPOSITION]
result = collectPatches(patches,img.shape[:2])
plt.imshow(result)
plt.title('preprocessed')
plt.show()


mydir = '1-1-1-1-1'#'3-3-3-3-3'
model_paths = [f for f in os.listdir(mydir) if os.path.isfile(os.path.join(mydir, f)) and f.startswith('moj_ulubiony_model') and f.endswith('.h5')]

for model_path in model_paths:
    model = load_model(MODEL_PATH)
    result = processImage(img,model)
    plt.imshow(result,cmap='gray')
    plt.title(model_path)
    plt.show()
```
