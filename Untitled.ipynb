```python
from RoadFinder import loadImage, preprocessorX,SIZE,IMPOSITION,SIDE,load_model
import matplotlib.pyplot as plt
import numpy as np
import cv2
import keras
import os
```

```python
img = loadImage("http://www.cs.toronto.edu/~vmnih/data/mass_roads/train/sat/10378765_15.tiff")
# img = cv2.resize(img,SIZE)
plt.imshow(img)
plt.show()
label = loadImage("http://www.cs.toronto.edu/~vmnih/data/mass_roads/train/map/10378765_15.tif")
# label = cv2.resize(label,SIZE)
plt.imshow(label)
plt.show()
```

```python
def collectPatches(patches,shape):
    acc = np.zeros(shape)
    rowlength = shape[0]//SIDE
    for i,patch in enumerate(patches):
        y0,x0 = SIDE*(i%rowlength),SIDE * (i//rowlength)
        acc[x0:x0+SIDE,y0:y0+SIDE] = patch
        
    return acc
        

def processImage(image,model):
    image = cv2.resize(image,SIZE)
    patches = np.array(preprocessorX(image))
    patches = patches.reshape(len(patches),SIDE+2*IMPOSITION,SIDE+2*IMPOSITION,1)
    
    mask1 = [np.all(patch==patch.mean()) for patch in patches]
    predictions = model.predict(patches)
#     for i in range(len(predictions)):
#         min1 = predictions[i].min()
#         d = predictions[i].max() - min1
#         predictions[i] -= min1
#         if d!=0:
#             predictions[i] /= d
    predictions[mask1] = predictions.min()
    predictions = predictions[:,IMPOSITION:-IMPOSITION,IMPOSITION:-IMPOSITION].reshape(len(predictions),SIDE,SIDE)
    result = collectPatches(predictions,image.shape[:2])
    
    min1 = result.min()
    d = result.max() - min1
    result -= min1
    if d!=0:
        result /= d
        
    result = cv2.resize(result,(1500,1500))
    result = cv2.GaussianBlur(result,(25,25),0)
    zx = 0.5
    result[result>zx] = 1
    result[result<=zx] = 0
#     result[result<np.percentile(result[result>0],94)] = 0
#     result[result!=0]=1
    kernel = np.ones((5,5),np.uint8)
    result = cv2.erode(result, kernel)
    kernel = np.ones((9,9),np.uint8)
    result = cv2.morphologyEx(result,cv2.MORPH_OPEN,kernel)
    kernel = np.ones((7,7),np.uint8)
    result = cv2.morphologyEx(result,cv2.MORPH_CLOSE,kernel)
    
    return result

patches = np.array(preprocessorX(cv2.resize(img, SIZE)))[:,IMPOSITION:-IMPOSITION,IMPOSITION:-IMPOSITION]
result1 = collectPatches(patches,img.shape[:2])
plt.imshow(result1,cmap='gray')
plt.title('preprocessed')
plt.show()


mydir = '.'#'1-1-1-1-1'#'3-3-3-3-3'
model_paths = [f for f in os.listdir(mydir) if os.path.isfile(os.path.join(mydir, f)) and f.startswith('moj_ulubiony_model') and f.endswith('.h5')]

for model_path in model_paths[-1:]:
    model = load_model(model_path)
    result2 = processImage(img,model)
    plt.imshow(result2,cmap='gray')
    plt.title(model_path)
    plt.show()
#     plt.hist(result2.ravel())
#     plt.title(model_path)
#     plt.show()
    
plt.imshow(np.abs(result1 - result2))
plt.title('gradient')
plt.show()

plt.imshow(np.abs(label - result2))
plt.title('gradient2')
plt.show()
```
